```python
def add_student():
    name=input("Enter name of student")
    marks=int(input("Enter marks of student"))
    student={}
    student["Name"]=name
    student["Marks"]=marks
    Record.append(student)
```

```python
def view_student():
    stu=input("Enter name of student you want to view")
    flag=0
    for student in Record:
        if student["Name"]==stu:
            print(student)
            flag=1
            break
    if flag==0:
        print(f"No record of: {stu}")
    
    
```
